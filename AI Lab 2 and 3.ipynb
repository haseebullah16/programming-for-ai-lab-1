# **FLOAT DATA TYPE **

# Standard Floating-Point Number:

```python
standard_float = 3.14
print(standard_float)  # Output: 3.14
```

# Scientific Notation:

```python
scientific_float = 2.5e-3
print(scientific_float)  # Output: 0.0025
```

# **Positive Infinity:**

```python
positive_infinity = float('inf')
print(positive_infinity)  # Output: inf
```

#**## Negative Infinity:**

```python
negative_infinity = float('-inf')
print(negative_infinity)  # Output: -inf
```

# TYPES CONVERSION

# **INTEGER TO FLOAT**

```python
integer_value = 42
float_value = float(integer_value)
print(float_value)  # Output: 42.0
```

# **FLOAT TO INTEGER**

```python
float_value = 3.14
integer_value = int(float_value)
print(integer_value)  # Output: 3


```

# **String to Integer:**

```python
string_value = "123"
integer_value = int(string_value)
print(integer_value)  # Output: 123
```

# **String to Float:**
python

```python
string_value = "3.14"
float_value = float(string_value)
print(float_value)  # Output: 3.14
```

# **BOOLEAN TO INTEGER**

```python
bool_value = True
int_value = int(bool_value)
print(int_value)  # Output: 1
```

# **INTEGER TO STRING**

```python
integer_value = 42
string_value = str(integer_value)
print(string_value)  # Output: '42'
```

# **FLOAT TO STRING**

```python
float_value = 3.14
string_value = str(float_value)
print(string_value)  # Output: '3.14'
```
